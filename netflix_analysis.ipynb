```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
```

```python
df = pd.read_csv("netflix_titles.csv")

df.head()
```

```python
df.info()
```

```python
df['director'] = df['director'].fillna("Unknown")
df['country'] = df['country'].fillna("Unknown")
df['cast'] = df['cast'].fillna("Unknown")
```

```python
sns.countplot(x='type', data=df)

plt.title("Movies vs TV Shows on Netflix")

plt.show()
```

```python
df['release_year'].value_counts().sort_index().plot(figsize=(10,5))

plt.title("Netflix Content Growth by Year")

plt.xlabel("Year")
plt.ylabel("Number of Titles")

plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar')

plt.title("Top Countries Producing Netflix Content")

plt.show()
```

```python
df['listed_in'].value_counts().head(10)
```

```python
df['listed_in'].value_counts().head(10).plot(kind='barh')

plt.title("Top 10 Genres on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")

plt.show()
```

```python
df['type'].value_counts(normalize=True) * 100
```

```python
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%')

plt.title("Movies vs TV Shows Distribution")

plt.ylabel("")

plt.show()
```

```python
df['director'].value_counts().head(10)
```

```python
df['director'].value_counts().head(10).plot(kind='bar')

plt.title("Top Directors on Netflix")

plt.show()
```

Key Insights

1. Netflix contains significantly more Movies than TV Shows.
2. Content production increased rapidly after 2015.
3. The United States produces the majority of Netflix titles.
4. Drama and International Movies are among the most common genres.
